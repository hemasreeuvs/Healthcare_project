```python
import pandas as pd

# Assuming your dataframe is named df
df = df.astype({
    'Hrs_RNDON': 'float64',
    'Hrs_RNDON_emp': 'float64',
    'Hrs_RNDON_ctr': 'float64',
    'Hrs_RNadmin': 'float64',
    'Hrs_RNadmin_emp': 'float64',
    'Hrs_RNadmin_ctr': 'float64',
    'Hrs_RN': 'float64',
    'Hrs_RN_emp': 'float64',
    'Hrs_RN_ctr': 'float64',
    'Hrs_LPNadmin': 'float64',
    'Hrs_LPNadmin_emp': 'float64',
    'Hrs_LPNadmin_ctr': 'float64',
    'Hrs_LPN': 'float64',
    'Hrs_LPN_emp': 'float64',
    'Hrs_LPN_ctr': 'float64',
    'Hrs_CNA': 'float64',
    'Hrs_CNA_emp': 'float64',
    'Hrs_CNA_ctr': 'float64',
    'Hrs_NAtrn': 'float64',
    'Hrs_NAtrn_emp': 'float64',
    'Hrs_NAtrn_ctr': 'float64',
    'Hrs_MedAide': 'float64',
    'Hrs_MedAide_emp': 'float64',
    'Hrs_MedAide_ctr': 'float64'
})
```

```python

df = df.astype({
    'PROVNUM': 'str',
    'PROVNAME': 'str',
    'CITY': 'str',
    'STATE': 'str',
    'COUNTY_NAME': 'str'
})
```

```python
print(df.dtypes)
```

```python
df['WorkDate'] = pd.to_datetime(df['WorkDate'], format='%m/%d/%Y', errors='coerce')
```

```python
print(df.dtypes)
```

```python
df.head(1)
```

Aggregration Calculations


```python


total_emp_hours = df['Hrs_RNDON_emp'].sum()
total_ctr_hours = df['Hrs_RNDON_ctr'].sum()


print(f"Total Employee Hours: {total_emp_hours}")
print(f"Total Contractor Hours: {total_ctr_hours}")
```

```python

total_emp_hours_all_states = df['Hrs_RNDON_emp'].sum()
total_ctr_hours_all_states = df['Hrs_RNDON_ctr'].sum()

#Grouping by State
state_totals = df.groupby('STATE').agg(
    total_emp_hours=('Hrs_RNDON_emp', 'sum'),
    total_ctr_hours=('Hrs_RNDON_ctr', 'sum')
)


state_totals['emp_percentage'] = (state_totals['total_emp_hours'] / total_emp_hours_all_states) * 100

state_totals['ctr_percentage'] = (state_totals['total_ctr_hours'] / total_ctr_hours_all_states) * 100


state_totals.reset_index(inplace=True)


print(state_totals[['STATE', 'emp_percentage', 'ctr_percentage']])
```

```python
import matplotlib.pyplot as plt

state_totals_sorted = state_totals.sort_values(by='emp_percentage', ascending=False)


plt.figure(figsize=(12, 6))


plt.plot(state_totals_sorted['STATE'], state_totals_sorted['emp_percentage'], marker='o', linestyle='-', color='#fd8d3c', label='Employee Hours %')


plt.plot(state_totals_sorted['STATE'], state_totals_sorted['ctr_percentage'], marker='s', linestyle='-', color='blue',label='Contractor Hours %')

plt.xticks(rotation=90)


plt.xlabel("State")
plt.ylabel("Percentage of Total Hours (%)")
plt.title("Employee vs Contractor Work Hour Percentages by State")


plt.legend()


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

roles = ['RNadmin', 'RN', 'LPNadmin', 'LPN', 'CNA', 'NAtrn', 'MedAide']


role_data = {}

for role in roles:
    emp_col = f'Hrs_{role}_emp'
    ctr_col = f'Hrs_{role}_ctr'


    total_emp_hours_all_states = df[emp_col].sum()
    total_ctr_hours_all_states = df[ctr_col].sum()


    state_totals = df.groupby('STATE').agg(
        total_emp_hours=(emp_col, 'sum'),
        total_ctr_hours=(ctr_col, 'sum')
    )


    state_totals['emp_percentage'] = (state_totals['total_emp_hours'] / total_emp_hours_all_states) * 100
    state_totals['ctr_percentage'] = (state_totals['total_ctr_hours'] / total_ctr_hours_all_states) * 100


    state_totals.reset_index(inplace=True)


    role_data[role] = state_totals


for role, data in role_data.items():
    print(f"Role: {role}")
    print(data[['STATE', 'emp_percentage', 'ctr_percentage']])
    print("\n")
```
